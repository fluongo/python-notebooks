```python
%matplotlib inline

import sys
import itertools
from repoze.lru import lru_cache

# Check the system type
if sys.platform == 'darwin':
    sys.path.append('/users/fluongo/documents/Python_stuff/subnets-ecog')
elif sys.platform == 'linux2':
    sys.path.append('/newhome/fluongo/python_modules/subnets-ecog')
    compSpec = '/Sohal'
else:
    sys.path.append('C:\\python_modules\\subnets-ecog')

import numpy as np
import matplotlib.pyplot as plt
import ecogRecordingParams as recParams
import os
import networkx as nx

import seaborn as sns
from pandas import DataFrame
import scipy.stats
from matplotlib.backends.backend_pdf import PdfPages
import scipy.stats as stats


def compute_zscore(data):
    data = np.nan_to_num(data)
    if len(data.shape) == 2:
        out = stats.mstats.zscore(data, axis = 1)
    else:
        out = stats.mstats.zscore(data)
    return out

def plot_im(data, refNames):
    pc = DataFrame(data, index = refNames, columns = refNames)
    sns.heatmap(pc, annot=False, fmt = '.2n', square = True, xticklabels = False)
    

def rsquared(x, y, nosquare = False):
    """ Return R^2 where x and y are array-like."""
    x = np.nan_to_num(x); y = np.nan_to_num(y)
    x = (x - np.mean(x))/np.std(x)
    y = (y - np.mean(y))/np.std(y)
    slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(x, y)
    if nosquare == True:
        output = r_value
    else:
        output = r_value**2
    
    return output

@lru_cache(maxsize = 40)
def load_data(patientID):
    data_loc = '/sohal-comp/fluongo/DURA_data/' + patientID + '/processedData/combined_10s/'
    data = np.load(data_loc + patientID + '_10s_proj_reduced.npy'); pc_data = data[()]

    coh_data = {}
    for band in ['delta', 'theta', 'alpha', 'beta', 'gamma']:
        coh_data[band] = np.load(data_loc + patientID + '_reducedcoh_10s_pr_' + band + '.npy')

    output = recParams.electrode_locs_fq_bands(patientID); brainRegions = output['brainRegions']; refNames = brainRegions.keys()
    
    return pc_data, coh_data, refNames

sns.set_style("white")
sns.set_style("ticks")
```

Now we will compute the R**2 values of the Zscored version of the coherence
Then fill them into a matrix which will give the maximal R2 i.e. which coherence value best predicts the projection

```python
pc_data, coh_data, refNames = load_data('EC71')

band = 'delta'
pc = pc_data['pc'][band]
cohs = coh_data[band]


f = plt.figure(figsize = [22, 22])

for PCrank in range(5):
    template = np.zeros_like(cohs[:,:,0])
    for i in range(cohs.shape[0]):
        for j in range(i,cohs.shape[0]):
            x = compute_zscore(cohs[i,j,:])
            y = compute_zscore(pc[PCrank,:])
            template[i,j] = rsquared(x , y)
            template[j,i] = template[i,j]
    plt.subplot(5,4,4*PCrank+1)
    plot_im(pc_data['comps']['pc'][band][:,:,PCrank], refNames); plt.title('original PC')
    plt.subplot(5,4,4*PCrank+2)
    plot_im(template, refNames); plt.title('R^2 values')
    plt.subplot(5,4,4*PCrank+3)
    compounded = pc_data['comps']['pc'][band][:,:,PCrank]*template
    plot_im(compounded, refNames); plt.title('PC weighted by the R^2, i.e. R^2 x PC')
    plt.subplot(5,4,4*PCrank+4)
    plot_im(np.round(np.abs(compounded)) >= 2*np.std(compounded.ravel()), refNames); plt.title('outliers of R^2 distribution')
```

After establishing these subsets of PCAs, we are interested in how much of the variance in the projection can be explained by just using these sections

- We can do this by simply convolving a PC with zeros everywhere except in those positions with the original PC
- We can also train a linear model fit using these dimensions as inputs and measure MSE

```python

def generate_R_singles(patientID, PCrank, band):
    # patientID: Just input the patient ID
    # PCrank: Which PC to plot
    # BAND: and what frequency band
    pc_data, coh_data, refNames = load_data(patientID)
    pc = pc_data['pc'][band]
    cohs = coh_data[band]
    
    
    all_R = []
    all_R2 = []


    pcComp = pc_data['comps']['pc'][band][:,:,PCrank]
    for i in range(pcComp.shape[0]):
        for j in range(i,pcComp.shape[0]):
            template = np.zeros_like(pcComp); template[i,j] = 1; template[j,i] = 1;
            subPC = pcComp*template
            curr_projection = np.dot(subPC.ravel(),cohs.reshape(cohs.shape[0]*cohs.shape[1], cohs.shape[2])); # Making the projection of the subPC onto the full
            all_R2.append(rsquared(compute_zscore(curr_projection), compute_zscore(pc[PCrank,:])))
            # Compute the R without square
            all_R.append(rsquared(compute_zscore(curr_projection), compute_zscore(pc[PCrank,:]), nosquare = True))
            #print 'R Squared value is ..... ' + str(all_R2[-1])
    print 'plot is for ' + patientID + ' and for PC number ' + str(PCrank) + ' in the ' + band + ' BAND'
    
    return all_R2, all_R

def generate_R_example(patientID, PCrank, band):
    # patientID: Just input the patient ID
    # PCrank: Which PC to plot
    # BAND: and what frequency band
    pc_data, coh_data, refNames = load_data(patientID)
    pc = pc_data['pc'][band]
    cohs = coh_data[band]
    thresh = 3 #How many standard deviations above to set the threshold

    f = plt.figure(figsize = [18, 4])

    template = np.zeros_like(cohs[:,:,0])
    for i in range(cohs.shape[0]):
        for j in range(i,cohs.shape[0]):
            x = compute_zscore(cohs[i,j,:])
            y = compute_zscore(pc[PCrank,:])
            template[i,j] = rsquared(x , y)
            template[j,i] = template[i,j]

    compounded = pc_data['comps']['pc'][band][:,:,PCrank]*template
    subPC = compounded*(np.round(np.abs(compounded)) >= thresh*np.std(compounded.ravel())) # Thresholding step
    curr_projection = np.dot(subPC.ravel(),cohs.reshape(cohs.shape[0]*cohs.shape[1], cohs.shape[2])); # Making the projection of the subPC onto the full

    for kk in range(10):
        plt.subplot(11,1,kk+1)
        plt.plot(y) # Plotting the Zscored original projection
        plt.plot(compute_zscore(curr_projection)); sns.despine(); plt.xlim([kk*3700, kk*3700 + 500]); plt.ylim([-3,3]); 
        plt.axis('off')
    
    f = plt.figure(figsize = [18, 4])
    plt.plot(y)
    plt.plot(compute_zscore(curr_projection)); sns.despine(); plt.ylim([-6,6]); 
    plt.axis('off')
    
    
    print 'plot is for ' + patientID + ' and for PC number ' + str(PCrank) + ' in the ' + band + ' BAND'
    print 'R Squared value is ..... ' + str(rsquared(compute_zscore(curr_projection), y))
    plt.figure()
    plot_im(subPC, refNames)
    
def plot_variance_v_R(patientID, PCrank, band):
    # This function will make a plot of variance as a function of R or Rsquared for each individual patient, band, and PCrank
    # patientID: Just input the patient ID
    # PCrank: Which PC to plot
    # BAND: and what frequency band
    pc_data, coh_data, refNames = load_data(patientID)
    pc = pc_data['pc'][band]
    cohs = coh_data[band]
    
    R2s, Rs = generate_R_singles(pID, 4, band)
    # Calculate the variances corresponding to each R2s and Rs
    
    variances = []
    
    for i in range(pcComp.shape[0]):
        for j in range(i,pcComp.shape[0]):
            variances.append(np.var(compute_zscore(cohs[i,j,:])))
    
    plt.scatter(variances, R2s)
    

pID = 'EC63b'
band = 'theta'

generate_R_example(pID, 1, band)
R2s, Rs = generate_R_singles(pID, 1, band)
plt.figure()
plt.hist(R2s, 20)
```

```python
def plot_top3_bottom12(patientID, band, plot_type = 'top'):
    X_all = []; Y_all = []; 
    for PCrank in range(5):
        R2s, names, cohs, names_pos, proj = plot_var_R_over_time(patientID, PCrank, band)
        # Find the top one and pull the name, position, etc...
        if plot_type == 'top':
            idx = R2s.argsort()[-3:]
        elif plot_type == 'all':
            idx = range(len(R2s))
        else:
            idx = R2s.argsort()[:len(R2s)/2]
        
        # Lists to hold the individual scatters
        top_names = '...'
        
        top_cohs = [];
        for top in idx:
            if plot_type == 'top':
                top_names += (names[top] + ' ')
                
            curr_coh = cohs[names_pos[top][0], names_pos[top][1], :]
            Xs, Ys = produce_X_Y(curr_coh, proj)
            X_all.extend(Xs); 
            Y_all.extend(Ys); 
            top_cohs.append(curr_coh);
        
        # Plot the raw signals
        if plot_type == 'top':
            if PCrank == 0:
                plt.figure(figsize = [20, 4])
            colors = ['r', 'g', 'b']; 
            for X, c in zip(top_cohs, colors):
                plt.subplot(5, 1, PCrank+1);plt.plot(compute_zscore(X), color = c)
            plt.title(top_names + 'for PC: ' + str(PCrank))
            plt.axis('off'); plt.xlim([15000, 15400]); plt.ylim([-3, 3])
    
    plt.figure()
    sns.jointplot(np.array(X_all), np.array(Y_all),kind = 'hex', xlim = [0,1], color="#4CB391", 
                  joint_kws={'gridsize':50, 'bins':'log'}); 
    plt.xlabel('R^2 with projection in 100 timesteps'); plt.ylabel('SD of Coherence in 100 timesteps')
    plt.suptitle(patientID + '...' + band + '...' + plot_type + top_names)
    plt.figure()
    sns.jointplot(np.array(X_all), np.array(Y_all),kind = 'reg',xlim = [0,1], color="#4CB391")
    plt.xlabel('R^2 with projection in 100 timesteps'); plt.ylabel('SD of Coherence in 100 timesteps')
    plt.suptitle(patientID + '...' + band + '...' + plot_type + top_names)


    
#plot_top3_bottom12('EC71', 'theta', plot_type = 'all')
plot_top3_bottom12('EC71', 'theta', plot_type = 'top')
#plot_top3_bottom12('EC71', 'theta', plot_type = 'bottom')
```
