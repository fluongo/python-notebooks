```python
%matplotlib inline

import sys
import itertools
from repoze.lru import lru_cache

# Check the system type
if sys.platform == 'darwin':
    sys.path.append('/users/fluongo/documents/Python_stuff/subnets-ecog')
elif sys.platform == 'linux2':
    sys.path.append('/newhome/fluongo/python_modules/subnets-ecog')
    compSpec = '/Sohal'
else:
    sys.path.append('C:\\python_modules\\subnets-ecog')

import numpy as np
import matplotlib.pyplot as plt
import ecogRecordingParams as recParams
import os
import networkx as nx

import seaborn as sns
from pandas import DataFrame
import scipy.stats
from matplotlib.backends.backend_pdf import PdfPages
import scipy.stats as stats


def compute_zscore(data):
    data = np.nan_to_num(data)
    if len(data.shape) == 2:
        out = stats.mstats.zscore(data, axis = 1)
    else:
        out = stats.mstats.zscore(data)
    return out

def plot_im(data, refNames):
    pc = DataFrame(data, index = refNames, columns = refNames)
    sns.heatmap(pc, annot=False, fmt = '.2n', square = True)
    

def rsquared(x, y, nosquare = False):
    """ Return R^2 where x and y are array-like."""
    x = np.nan_to_num(x); y = np.nan_to_num(y)
    x = (x - np.mean(x))/np.std(x)
    y = (y - np.mean(y))/np.std(y)
    slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(x, y)
    if nosquare == True:
        output = r_value
    else:
        output = r_value**2
    
    return output

@lru_cache(maxsize = 40)
def load_data(patientID):
    if sys.platform == 'darwin':
        data_loc = '/Volumes/sohal-comp/DURA_data/' + patientID + '/processedData/combined_10s/'
    elif sys.platform == 'linux2':
        data_loc = '/sohal-comp/fluongo/DURA_data/' + patientID + '/processedData/combined_10s/'
    data = np.load(data_loc + patientID + '_10s_proj_reduced.npy'); pc_data = data[()]

    coh_data = {}
    for band in ['delta', 'theta', 'alpha', 'beta', 'gamma']:
        coh_data[band] = np.load(data_loc + patientID + '_reducedcoh_10s_pr_' + band + '.npy')

    output = recParams.electrode_locs_fq_bands(patientID, shortNames = True); brainRegions = output['brainRegions']; refNames = brainRegions.keys()
    
    return pc_data, coh_data, refNames

sns.set_style("white")
sns.set_style("ticks")
```

```python
pc_data, coh_data, refNames = load_data('EC79')

band = 'theta'
pc = pc_data['pc'][band]
cohs = coh_data[band]


f = plt.figure(figsize = [22, 28])

for PCrank in range(5):
    template = np.zeros_like(cohs[:,:,0])
    for i in range(cohs.shape[0]):
        for j in range(i,cohs.shape[0]):
            x = compute_zscore(cohs[i,j,:])
            y = compute_zscore(pc[PCrank,:])
            template[i,j] = rsquared(x , y)
            template[j,i] = template[i,j]
    plt.subplot(5,4,4*PCrank+1)
    plot_im(pc_data['comps']['pc'][band][:,:,PCrank], refNames); plt.title('original PC')
    plt.subplot(5,4,4*PCrank+2)
    plot_im(template, refNames); plt.title('R^2 values')
    plt.subplot(5,4,4*PCrank+3)
    compounded = pc_data['comps']['pc'][band][:,:,PCrank]*template
    plot_im(compounded, refNames); plt.title('PC weighted by the R^2, i.e. R^2 x PC')
    plt.subplot(5,4,4*PCrank+4)
    plot_im(template >= 3*np.std(template.ravel()), refNames); plt.title('outliers of R^2 distribution')
```

```python
def count_nonzero_PC(PC):
    # Function counts the nonzero elements present in a PC
    idx = np.where(PC != 0)
    count = 0
    for i,j in zip(idx[0], idx[1]):
        if i == j:
            count+=1
        if i != j:
            count += 0.5
    count = int(count)
    total = (PC.shape[0]**2 - PC.shape[0])/2 + PC.shape[0]
    return count, total
    
def generate_R_singles(patientID, PCrank, band):
    # patientID: Just input the patient ID
    # PCrank: Which PC to plot
    # BAND: and what frequency band
    pc_data, coh_data, refNames = load_data(patientID)
    pc = pc_data['pc'][band]
    cohs = coh_data[band]
    
    
    all_R = []
    all_R2 = []


    pcComp = pc_data['comps']['pc'][band][:,:,PCrank]
    for i in range(pcComp.shape[0]):
        for j in range(i,pcComp.shape[0]):
            template = np.zeros_like(pcComp); template[i,j] = 1; template[j,i] = 1;
            subPC = pcComp*template
            curr_projection = np.dot(subPC.ravel(),cohs.reshape(cohs.shape[0]*cohs.shape[1], cohs.shape[2])); # Making the projection of the subPC onto the full
            all_R2.append(rsquared(compute_zscore(curr_projection), compute_zscore(pc[PCrank,:])))
            # Compute the R without square
            all_R.append(rsquared(compute_zscore(curr_projection), compute_zscore(pc[PCrank,:]), nosquare = True))
            #print 'R Squared value is ..... ' + str(all_R2[-1])
    print 'plot is for ' + patientID + ' and for PC number ' + str(PCrank) + ' in the ' + band + ' BAND'
    
    return all_R2, all_R

def generate_R_example(patientID, PCrank, band, thresh = 3, showPlot = True):
    # patientID: Just input the patient ID
    # PCrank: Which PC to plot
    # BAND: and what frequency band
    # Thresh: How many standard deviations above to set the threshold
    pc_data, coh_data, refNames = load_data(patientID)
    pc = pc_data['pc'][band]
    cohs = coh_data[band]
    
    template = np.zeros_like(cohs[:,:,0])
    for i in range(cohs.shape[0]):
        for j in range(i,cohs.shape[0]):
            x = compute_zscore(cohs[i,j,:])
            y = compute_zscore(pc[PCrank,:])
            #x = cohs[i,j,:]
            #y = pc[PCrank, :]
            template[i,j] = rsquared(x , y)
            template[j,i] = template[i,j]
    
    # Thresholding is carried out on the R^2s then leaves the original loadings...
    compounded = pc_data['comps']['pc'][band][:,:,PCrank]*(template >= thresh*np.std(template.ravel())) # Thresholding step
    subPC = compounded
    
    curr_projection = np.dot(subPC.ravel(),cohs.reshape(cohs.shape[0]*cohs.shape[1], cohs.shape[2])); # Making the projection of the subPC onto the full
    cnt, tot = count_nonzero_PC(subPC)

    if showPlot == True:
        print 'USING: ' +  str(cnt)+ ' out of ' + str(tot)
        print 'plot is for ' + patientID + ' and for PC number ' + str(PCrank) + ' in the ' + band + ' BAND'
        print 'R Squared value is ..... ' + str(rsquared(compute_zscore(curr_projection), y))
    
        f = plt.figure(figsize = [18, 8])

        plt.subplot(11,1,1); 
        plt.plot(1, label = 'PC projection (Zscored)'); plt.plot(1, label = 'Linear projection of subset (Zscored)')
        plt.axis('off'); plt.legend()
        for kk in range(10):
            plt.subplot(11,1,kk+2)
            plt.plot(y) # Plotting the Zscored original projection
            plt.plot(compute_zscore(curr_projection)); 
            sns.despine(); plt.xlim([kk*3700, kk*3700 + 500]); plt.ylim([-2,2]); 
            plt.axis('off')
        plt.suptitle('R Squared value is ..... ' + str(rsquared(compute_zscore(curr_projection), y)))
        
        
        f = plt.figure(figsize = [18, 4])
        plt.plot(y, label = 'PC projection (Zscored)')
        plt.plot(compute_zscore(curr_projection), label = 'Linear projection of subset (Zscored)'); sns.despine(); plt.ylim([-6,6]); 
        plt.axis('off')
        plt.title('R Squared value is ..... ' + str(rsquared(compute_zscore(curr_projection), y)))
        plt.legend()
        plt.figure()
        plot_im(subPC, refNames)
        
    return rsquared(compute_zscore(curr_projection), y), cnt

pID = 'EC79'
band = 'theta'
PCrank = 4

r, cnt = generate_R_example(pID, PCrank, band, thresh = 1.5)
R2s, Rs = generate_R_singles(pID, PCrank, band)
plt.figure()
#plt.hist(R2s, 20)
```

```python
@lru_cache(maxsize = 100)
def plot_nodes_vs_R(pID, PCrank, band):
    Rs = []; cnts = [] ;
    for thresh in np.linspace(0,5,15):
        r, cnt = generate_R_example(pID, PCrank, band, thresh, showPlot = False)
        if np.isnan(r):
            Rs.append(0)
        else:
            Rs.append(r)
        cnts.append(cnt)
    return Rs, cnts

pID = 'EC71'
PCrank = 0
band = 'theta'


R_all = {}; C_all = {}
for pID in ['EC63b', 'EC71', 'EC72', 'EC77', 'EC79', 'EC80']:
    print pID
    R_all[pID] =[]; C_all[pID] = []
    for band in ['theta', 'delta', 'beta', 'alpha']:
        for PCrank in range(5):
            Rs, cnts = plot_nodes_vs_R(pID, PCrank, band)
            R_all[pID].append(Rs)
            C_all[pID].append(cnts)
```
