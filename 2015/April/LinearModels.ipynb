```python
%matplotlib inline

import sys
import itertools
from repoze.lru import lru_cache

# Check the system type
if sys.platform == 'darwin':
    sys.path.append('/users/fluongo/documents/Python_stuff/subnets-ecog')
elif sys.platform == 'linux2':
    sys.path.append('/home/fluongo/python_modules/subnets-ecog')
    compSpec = '/Sohal'
else:
    sys.path.append('C:\\python_modules\\subnets-ecog')

import numpy as np
import matplotlib.pyplot as plt
import ecogRecordingParams as recParams
import os
import networkx as nx

import seaborn as sns
from pandas import DataFrame
import scipy.stats
from matplotlib.backends.backend_pdf import PdfPages
import scipy.stats as stats
from repoze.lru import lru_cache

sns.set_style("white")
sns.set_style("ticks")

@lru_cache(maxsize = 200)
def load_data(patientID):
    data_loc = '/sohal-comp/fluongo/DURA_data/' + patientID + '/processedData/combined_10s/'
    data = np.load(data_loc + patientID + '_10s_proj_reduced.npy'); pc_data = data[()]

    coh_data = {}
    for band in ['delta', 'theta', 'alpha', 'beta', 'gamma']:
        coh_data[band] = np.load(data_loc + patientID + '_reducedcoh_10s_pr_' + band + '.npy')

    output = recParams.electrode_locs_fq_bands(patientID, shortNames = True); brainRegions = output['brainRegions']; refNames = brainRegions.keys()
    
    return pc_data, coh_data, refNames

def plot_scatter_grid(Xs, Ys):
    sns.jointplot(np.array(Xs), np.array(Ys),kind = 'hex', xlim = [0,1], color="#bf9005", 
                  joint_kws={'gridsize':50, 'bins':'log'})
    
def rsquared(x, y, nosquare = False):
    """ Return R^2 where x and y are array-like."""
    x = np.nan_to_num(x); y = np.nan_to_num(y)
    x = (x - np.mean(x))/np.std(x)
    y = (y - np.mean(y))/np.std(y)
    slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(x, y)
    if nosquare == True:
        output = r_value
    else:
        output = r_value**2
    
    return output  

def compute_zscore(data):
    data = np.nan_to_num(data)
    if len(data.shape) == 2:
        out = stats.mstats.zscore(data, axis = 1)
    else:
        out = stats.mstats.zscore(data)
    return out

def plot_curved_bin(Xs, Ys, bins = 1000, constantBinSize = False, constantSize = 2000):
    

    if constantBinSize == True:
        idx = np.argsort(Xs)
        Xsort = Xs[idx]
        Ysort = Ys[idx]
        newX = []; newY = []
        nn = 0
        while nn < len(Xsort):
            newX.append(np.mean(Xsort[nn:nn+constantSize]))
            newY.append(np.mean(Ysort[nn:nn+constantSize]))
            nn += constantSize
    else:
        res = np.abs(np.max(Xs)-np.min(Xs))/bins
        newX = []; newY = []
        for center in np.linspace(np.min(Xs), np.max(Xs), bins):
            newX.append(center); 
            newY.append(np.mean(Ys[(Xs > center - res/2) & (Xs < center + res/2)  ]))

    
        
    plt.plot(newX, newY)
    plt.plot([np.min(newX), np.max(newX)], [np.min(newX), np.max(newX)], linestyle = '--')
```

```python
from sklearn import cross_validation
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

pc, cohs, refNames = load_data('EC71')

PCrank = 1
offset = 1

Xs = []; Ys = []
for i in range(pc['pc']['theta'].shape[1]-offset):
    Xs.append(pc['pc']['theta'][PCrank, i])
    Ys.append(pc['pc']['theta'][PCrank, i+offset])
# Clean up NaNs
Xs = np.nan_to_num(np.array(Xs))
Ys = np.nan_to_num(np.array(Ys))

errors = []
for n in range(10):
    X_train, X_test, y_train, y_test = cross_validation.train_test_split(Xs, Ys, test_size=0.2)
    lr = LinearRegression()
    lr.fit(X_train[:, np.newaxis], y_train)
    errors.append(mean_squared_error(y_test, lr.predict(X_test[:, np.newaxis])))

plt.figure()
plt.hexbin(Xs, Ys, bins = 'log', gridsize = 50)

plot_curved_bin(Xs, Ys, constantBinSize = False)
plt.figure()
plot_curved_bin(Xs, Ys, constantBinSize = True, constantSize = 500)
```
