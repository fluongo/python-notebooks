```python
%matplotlib qt
import seaborn as sns
from pandas import DataFrame

band = 'theta'
patientID = 'EC63'
output = recParams.electrode_locs_fq_bands(patientID)
brainRegions = output['brainRegions']
refNames = brainRegions.keys()
interval = '30s'
data = np.load('Y:\\DURA_data\\'+ patientID + '\\processedData\\combined_' + interval + '\\PCA_all_reduced.npy')

def plot_im(data):
    pc = DataFrame(data, index = refNames, columns = refNames)
    sns.heatmap(pc, annot=False, fmt = '.2n', square = True, xticklabels = False, center = 0)
    
def sparsify(data):
    # Go through all possible thresholds
    costs = []
    for thr in np.arange(50):
        thresh1 = np.percentile(data, 100-thr)
        thresh2 = np.percentile(data, thr)
        sp1 = 1*(data > thresh1); 
        sp2 = -1*(data < thresh2); 
        sparse_data_temp = sp1+sp2
        dot = np.sum(data*sparse_data_temp)
        costs.append(dot - np.sum(abs(sparse_data_temp)))
    
    idx = np.argmax(costs) # finds the index of maxvalue
    #print idx
    
    #There is a weird bug where it falls exactly on the amount resulting in not mirror symmetrical results
    thresh1 = np.percentile(data, 100-idx)-.01; 
    thresh2 = np.percentile(data, idx)+.01; 
    
    sp1 = 1*(data > thresh1); 
    sp2 = -1*(data < thresh2); 
    sparse_data = sp1+sp2;
    
    #Now prune the cases where it might cost more to make a move...
    orig_cost = costs[idx]
    sp_pen = np.sum(abs(sparse_data)) - 1
    pcS_cost = np.zeros_like(sparse_data, dtype=np.float)

    for X, Y in zip(np.nonzero(sparse_data)[0], np.nonzero(sparse_data)[1]):
        pcS_temp = np.copy(sparse_data)
        pcS_temp[X, Y] = 0  # Change one of them
        pcS_cost[X, Y] = abs(orig_cost - (np.sum(data * pcS_temp) - sp_pen))
    
    # Now do the pruning
    weed_thr = np.std(pcS_cost.ravel())
    sparse_data = sparse_data*(pcS_cost>weed_thr)

    return sparse_data
for patientID in ['EC63', 'EC71', 'EC72', 'EC74', 'EC77']:
    output = recParams.electrode_locs_fq_bands(patientID); brainRegions = output['brainRegions']; refNames = brainRegions.keys()
    data = np.load('Y:\\DURA_data\\'+ patientID + '\\processedData\\combined_30s\\PCA_all_reduced.npy')
    
    plt.figure(figsize = [22, 17])
    sp_ind = 1;
    for band in ['delta', 'alpha', 'theta', 'beta', 'gamma', 'gammaHi']:
        for nRank in [0, 1, 2, 3, 4]:
            plt.subplot(6,5,sp_ind); 
            plot_im(sparsify(data[()][band][:,:,nRank])) 
            if nRank == 2:
                plt.title(band)
            sp_ind += 1
    plt.suptitle(patientID)
    #plt.savefig('C:\\figure_export\\allPC_sparse_' + patientID + '.png')
```

This code is meant to go through and make a determination whether there is a positive or negative weight associated with a certain area

```python
# Creating the template which is 11 x 11 x 4 patients (EC63, EC71, EC72, EC77)

reload(recParams)

refNames_template = ['ant OFC', 'pos OFC', 'ant ST', 'mid ST', 'pos ST', 
                     'inf CIN', 'sup CIN', 'ant INS', 'pos INS', 'Amygdala', 'Hippocampus']

for nRank in [0,1,2]:
    subplot_pos = 1
    plt.figure(figsize = [22,17])

    for band in ['delta', 'theta', 'alpha', 'beta', 'gamma', 'gammaHi']:
        kk = 0; # patientID counter
        X = np.zeros([11,11,4])
        counts = np.zeros([11,11])

        for patientID in ['EC63b', 'EC71', 'EC72', 'EC77']:
            output = recParams.electrode_locs_fq_bands(patientID); brainRegions = output['brainRegions']; refNames = brainRegions.keys()
            data = np.load('Y:\\DURA_data\\'+ patientID + '\\processedData\\combined_30s\\PCA_all_reduced.npy')
            curr_pc = data[()][band][:,:,nRank]
            curr_pcS = sparsify(curr_pc)
            #plt.figure(); sns.heatmap(DataFrame(curr_pcS))
            for area in refNames:
                
                #########################
                # Section for dealing with OFC into pos ant OFC
                
                if area == 'OFC':
                    areaOG = area
                    for area in ['pos OFC', 'ant OFC']:
                        for area2 in refNames:
                            if area2 in refNames_template:
                                idx = refNames_template.index(area)
                                idx2 = refNames_template.index(area2)
                                if abs(curr_pcS[refNames.index(areaOG), refNames.index(area2)]) == 1:
                                    X[idx, idx2, kk] += 1
                                    X[idx2, idx, kk] += 1 # Add 1 every time
                                counts[idx, idx2] += 1
                                counts[idx2, idx] += 1
                            else:
                                continue
                    area = 'jibber' #At the end reset area to something that will continue
                
                # Section for dealing with INS into pos ant INS
                if area == 'INS':
                    areaOG = area
                    for area in ['pos INS', 'ant INS']:
                        for area2 in refNames:
                            if area2 in refNames_template:
                                idx = refNames_template.index(area)
                                idx2 = refNames_template.index(area2)
                                if abs(curr_pcS[refNames.index(areaOG), refNames.index(area2)]) == 1:
                                    X[idx, idx2, kk] += 1
                                    X[idx2, idx, kk] += 1 # Add 1 every time
                                counts[idx, idx2] += 1
                                counts[idx2, idx] += 1
                            else:
                                continue
                    area = 'jibber' #At the end reset area to something that will continue
                
                
                #############################
                
                if area in refNames_template:
                    for area2 in refNames:
                        if area2 in refNames_template:
                            idx = refNames_template.index(area)
                            idx2 = refNames_template.index(area2)
                            X[idx, idx2, kk] = abs(curr_pcS[refNames.index(area), refNames.index(area2)]) # Substitute with 1 if its there
                            counts[idx, idx2] += 1
                        else:
                            continue
                else:
                    continue
            kk+=1

        plt.subplot(3,3,subplot_pos); subplot_pos += 1
        template = DataFrame(np.sum(X, axis = 2)/counts, index = refNames_template, columns = refNames_template)
        sns.heatmap(template, annot=True, fmt = '.2n', square = True, xticklabels = False,  cbar = False); plt.title(band)

    plt.subplot(3,3,8); 
    sns.heatmap(DataFrame(counts, index = refNames_template, columns = refNames_template),
                xticklabels = False, annot = True, square = True,  cbar = False); 
    plt.title('total counts possible out of 4 patients (63, 71, 72, 77)')

    plt.suptitle('PC ' + str(nRank))
```
