```python
%matplotlib qt
import seaborn as sns
from pandas import DataFrame

band = 'theta'
patientID = 'EC63'
output = recParams.electrode_locs_fq_bands(patientID)
brainRegions = output['brainRegions']
refNames = brainRegions.keys()
interval = '30s'

############### Computer specific part

if sys.platform == 'darwin':
    compSpec = '/Volumes/sohal-comp'
else:
    compSpec = 'Y:'

data = np.load(compSpec + os.sep + 'DURA_data' + os.sep + patientID + os.sep + 'processedData' + os.sep + 
               'combined_' + interval + os.sep + 'PCA_all_reduced.npy')

################

def plot_im(data):
    pc = DataFrame(data, index = refNames, columns = refNames)
    sns.heatmap(pc, annot=False, fmt = '.2n', square = True, xticklabels = False, center = 0)
    
def sparsify(data):
    # Go through all possible thresholds
    costs = []
    for thr in np.arange(50):
        thresh1 = np.percentile(data, 100-thr)
        thresh2 = np.percentile(data, thr)
        sp1 = 1*(data > thresh1); 
        sp2 = -1*(data < thresh2); 
        sparse_data_temp = sp1+sp2
        dot = np.sum(data*sparse_data_temp)
        costs.append(dot - np.sum(abs(sparse_data_temp)))
    
    idx = np.argmax(costs) # finds the index of maxvalue
    #print idx
    
    #There is a weird bug where it falls exactly on the amount resulting in not mirror symmetrical results
    thresh1 = np.percentile(data, 100-idx)-.01; 
    thresh2 = np.percentile(data, idx)+.01; 
    
    sp1 = 1*(data > thresh1); 
    sp2 = -1*(data < thresh2); 
    sparse_data = sp1+sp2;
    
    #Now prune the cases where it might cost more to make a move...
    orig_cost = costs[idx]
    sp_pen = np.sum(abs(sparse_data)) - 1
    pcS_cost = np.zeros_like(sparse_data, dtype=np.float)

    for X, Y in zip(np.nonzero(sparse_data)[0], np.nonzero(sparse_data)[1]):
        pcS_temp = np.copy(sparse_data)
        pcS_temp[X, Y] = 0  # Change one of them
        pcS_cost[X, Y] = abs(orig_cost - (np.sum(data * pcS_temp) - sp_pen))
    
    # Now do the pruning
    weed_thr = np.std(pcS_cost.ravel())
    sparse_data = sparse_data*(pcS_cost>weed_thr)

    return sparse_data


##########################################################
#  Notes about how this function deals with OFC and INS
#
# Coherence within INS translates to coherence within ant INS and pos INS but not between pos and ant INS
# Coherence between INS and region X corresponds to coherence between both ant INS and pos INS with region X
#
# Same applys to OFC
##########################################################

def project_onto_template(curr_pcS, refNames, refNames_template):
    
    projected_onto_template = np.zeros([len(refNames_template),len(refNames_template)])
    counts = np.zeros([len(refNames_template),len(refNames_template)])
    
    for area in refNames:                
        #########################
        # Section for dealing with OFC into pos ant OFC
        if area == 'OFC':
            areaOG = area
            # Check for the identity (i.e. set the diagonal)
            for area in ['pos OFC', 'ant OFC']:
                idx = refNames_template.index(area)
                projected_onto_template[idx, idx] = curr_pcS[refNames.index(areaOG), refNames.index(areaOG)]
                counts[idx, idx] = 1 #iterate count
            
            #Check everything else
            for area in ['pos OFC', 'ant OFC']:
                for area2 in refNames:
                    if area2 in refNames_template:
                        idx = refNames_template.index(area)
                        idx2 = refNames_template.index(area2)
                        if abs(curr_pcS[refNames.index(areaOG), refNames.index(area2)]) == 1:
                            projected_onto_template[idx, idx2] = curr_pcS[refNames.index(areaOG), refNames.index(area2)]
                            projected_onto_template[idx2, idx] = projected_onto_template[idx, idx2]
                            counts[idx, idx2] = 1 #iterate count
                            counts[idx2, idx] = 1 #iterate count
                        else:
                            counts[idx, idx2] = 1 #iterate count
                            counts[idx2, idx] = 1 #iterate count
                            continue
            area = 'jibber' #At the end reset area to something that will continue

        # Section for dealing with INS into pos ant INS
        if area == 'INS':
            areaOG = area
            # Check for the identity (i.e. set the diagonal)
            for area in ['pos INS', 'ant INS']:
                idx = refNames_template.index(area)
                projected_onto_template[idx, idx] = curr_pcS[refNames.index(areaOG), refNames.index(areaOG)]
                counts[idx, idx] = 1 #iterate count

            
            # Check for everything else
            for area in ['pos INS', 'ant INS']:
                for area2 in refNames:
                    if area2 in refNames_template:
                        idx = refNames_template.index(area)
                        idx2 = refNames_template.index(area2)
                        if abs(curr_pcS[refNames.index(areaOG), refNames.index(area2)]) == 1:
                            projected_onto_template[idx, idx2] = curr_pcS[refNames.index(areaOG), refNames.index(area2)]
                            projected_onto_template[idx2, idx] = projected_onto_template[idx, idx2]
                            counts[idx, idx2] = 1 #iterate count
                            counts[idx2, idx] = 1 #iterate count
                        else:
                            counts[idx, idx2] = 1 #iterate count
                            counts[idx2, idx] = 1 #iterate count
                            continue
            area = 'jibber' #At the end reset area to something that will continue

        #############################

        if area in refNames_template:
            for area2 in refNames:
                if area2 in refNames_template:
                    idx = refNames_template.index(area)
                    idx2 = refNames_template.index(area2)
                    projected_onto_template[idx, idx2] = curr_pcS[refNames.index(area), refNames.index(area2)] # Substitute with 1 if its there
                    counts[idx, idx2] += 1
                else:
                    continue
        else:
            continue
            
    return projected_onto_template, counts
        
```

### Counting Motifs the way that vikaas suggested

```python
import itertools

# Function for counting motifs in a given PC

def count_motifs(pcS, refNames_template, nRegions):
    indices = np.nonzero(pcS)
    nz_list = []; 
    # Making the list of indices and only making it half
    for X, Y in zip(indices[0], indices[1]):
        if X - Y >= 0:
            nz_list.append([X, Y])
    
    all_motifs = []
    all_values = []
    
    perm_list = list(itertools.combinations(range(len(nz_list)), nRegions))
    for X in perm_list:
        curr_idx = []
        for i in range(nRegions):
            curr_idx.append(nz_list[X[i]])
            
        all_motifs.append(curr_idx)
        # Compile the values
        vals = [pcS[tuple(X)] for X in curr_idx]
        all_values.append(vals)
        
    return all_motifs, all_values

def make_unique_lists(all_M, all_V): #This is not super fast but it will do for now, just needs to be split on multiple processors

    unique_M = []
    unique_V = []

    cnt = -1
    counts = []

    for X, V in zip(M_all, V_all):
        cnt += 1
        didNotSet = False
        if X in unique_M:
            # Check all indices where it is in unique_M
            indices = [i for i, x in enumerate(unique_M) if x == X]
            for i in indices:
                if V_all[i] == V or np.allclose(array(V_all[i]),-array(V)):
                    counts[i] += 1
                    didNotSet = True
        elif didNotSet == False:
            counts.append(1)
            unique_M.append(X)
            unique_V.append(V_all[cnt])
        
    return unique_M, unique_V, counts

# Start with a seed region (one column of sparse matrix)
# Then count all regions of 3, 4, 5, 6, 7, 8, 9 that do not include the seed region



# Do this for all patients across all bands and then reiterate through all patients and all bands
# Also normalize by how many times it could have come up (i.e. how many of the patients had all of the regions necessary to satisfy)

band = sys.argv[1]
all_motifs_data = {}

for nRegions in [3,4,5,6,7,8]:
    M_all = []
    V_all = []
    for nRank in [0, 1, 2, 3, 4]:
        for patientID in ['EC63b', 'EC71', 'EC72', 'EC77', 'EC79']:
            output = recParams.electrode_locs_fq_bands(patientID); brainRegions = output['brainRegions']; refNames = brainRegions.keys()
            data = np.load(compSpec + os.sep + 'DURA_data' + os.sep + patientID + os.sep + 'processedData' + os.sep + 
               'combined_' + interval + os.sep + 'PCA_all_reduced.npy')
            curr_pc = data[()][band][:,:,nRank]
            curr_pcS = sparsify(curr_pc)
            template_proj, counts_current = project_onto_template(curr_pcS, refNames, refNames_template)
            M, V = count_motifs(template_proj, refNames_template, nRegions)

            M_all.extend(M)
            V_all.extend(V)

    unique_M, unique_V, M_counts = make_unique_lists(M_all, V_all)
    
    all_motifs_data[str(nRegions)] = {}
    all_motifs_data[str(nRegions)]['motifs'] = unique_M
    all_motifs_data[str(nRegions)]['values'] = unique_V
    all_motifs_data[str(nRegions)]['counts'] = M_counts
    
    np.save('/home/fluongo/output_data/motifs_'+band+'.npy', all_motifs_data)
```
