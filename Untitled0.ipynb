```python

dataTemp = data[2,2,:]
plt.plot(dataTemp)
baseline = np.zeros_like(dataTemp)
X = np.linspace(-1,1,1000)
nn = 2

while nn < len(dataTemp)-2:
    
    if nn % 1000 == 0:
        print nn
        
    if nn + 500 > len(dataTemp):
        kde = stats.gaussian_kde(dataTemp[nn:])
        
    elif nn - 500 <= 0:
        kde = stats.gaussian_kde(dataTemp[nn:])
        
    else:
        kde = stats.gaussian_kde(dataTemp[nn-500:nn+500])
    nn += 1
    pdf = kde.evaluate(X)
    baseline[nn] = X[np.argmax(pdf)]
    
```

```python
plt.plot(dataTemp)
plt.plot(baseline, color = 'r')
plt.figure()
plt.plot(dataTemp - baseline, color = 'g' )
plt.figure()
rVar = running_var(dataTemp, 100)
plt.plot(rVar)
```
