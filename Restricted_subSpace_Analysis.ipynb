This is code that will perform the subspace analysis, it can find the top 9 motifs of a given frequency band then load the reference matrix for a given dataset and plot some KDE of the distribution of a 3d entity in 1 column as 3 2d plots

```python

%matplotlib qt


import seaborn as sns
import sys

# Check the system type
if sys.platform == 'darwin':
    sys.path.append('/users/fluongo/documents/Python_stuff/subnets-ecog')
else:
    sys.path.append('C:\\python_modules\\subnets-ecog')

import numpy as np
import matplotlib.pyplot as plt
import ecogRecordingParams as recParams
import os
import seaborn as sns
from pandas import DataFrame

############### Computer specific part

if sys.platform == 'darwin':
    compSpec = '/Volumes/sohal-comp'
else:
    compSpec = 'C:\\SUBNETS_localData'

################

data_fn = compSpec + os.sep + 'output_data' + os.sep + 'motifs_theta.npy'
 
refNames_template = ['ant OFC', 'pos OFC', 'ant ST', 'mid ST', 'pos ST', 
                     'inf CIN', 'sup CIN', 'ant INS', 'pos INS', 'Amygdala', 'Hippocampus']

```

```python

def plot_coords(coords, vals):
    X = np.zeros([len(refNames_template), len(refNames_template)])
    
    for kk, V in zip(coords, vals):
        X[tuple(kk)] = V
        X[tuple(np.sort(kk))] = V   
    template = DataFrame(X, index = refNames_template, columns = refNames_template)
    sns.heatmap(template, annot = False, fmt = '.2n', square = True, xticklabels = False,  cbar = False, center = 0); 
    
band = 'delta'
nM = '3'
data_fn = compSpec + os.sep + 'output_data' + os.sep + 'disV_motifs_' + band + '.npy'
data = np.load(data_fn)

unique_M = data[()][nM]['motifs']
unique_V = data[()][nM]['values'] 
counts = data[()][nM]['counts']

plt.figure(figsize = [20,4])
kk = 1
for X in data[()].keys()[0:4]:
    plt.subplot(1,4,kk); kk+=1
    countsT = data[()][X]['counts']
    plt.hist(countsT, bins = np.arange(max(countsT)+2)+0.5); plt.yscale('log'); plt.title(X + ' region Motifs')
    
# Code below will plot the top 10 non-diagonal

indices = sorted(range(len(counts)), key=lambda i: counts[i])[-900:]
plt.figure(figsize = [18,18])
kk = 1; 
top9_noD = []; top9 = []
nn = 1
for X in reversed(indices):
    
    if nn < 10:
        top9.append(unique_M[X]); nn+=1
        
    if sum(np.diff(np.array(unique_M[X]))) != 0: # This enrichs for at least one non-diagonal interaction
        plt.subplot(3,3,kk); kk+=1
        plot_coords(unique_M[X], unique_V[X])
        top9_noD.append(unique_M[X])
        plt.title('occurred ' + str(counts[X]) + ' times')
    if kk == 10:
        break
```

```python
# Pull a patients full coherence for these areas

#Redefine the motifs using the names with nested list

top9_motifs_regions = []

for motif in top9_noD:
    temp = []
    for interaction in motif:
        temp.append([refNames_template[interaction[0]], refNames_template[interaction[1]]])
    top9_motifs_regions.append(temp)


top9_motifs_regions
```

###### This code generates the datasets of the 9 motifs from the above DataT

```python


for motif in top9_motifs_regions:
    kk = 0
    temp = np.zeros([3,dataT.shape[2]])
    for regions in motif:
        temp[kk, :] = dataT[refNames_template.index(regions[0]), refNames_template.index(regions[1]), :]
        temp[kk, :] = temp[kk, :] - mean(temp[kk, :])
        kk += 1
    plt.figure()

    plt.plot(temp.T)
```

```python
# Plotting using seaborn joint plot
%pylab qt

from scipy.stats import kendalltau
plt.figure(figsize = [22, 12])

plot_num = 1
for motif in top9_motifs_regions[0:6]:
    kk = 0
    temp = np.zeros([3,dataT.shape[2]])
    
    # Grab coherence values for the motif
    for regions in motif:
        temp[kk, :] = dataT[refNames_template.index(regions[0]), refNames_template.index(regions[1]), :]
        temp[kk, :] = temp[kk, :] - mean(temp[kk, :])
        kk += 1
    
    # Now generate the plot
    for sets in [[0,1], [0,2], [1,2]]:

        X = temp[sets[0],:]
        Y = temp[sets[1],:]

        #sns.jointplot(X, Y, kind="hex", color="#4CB391", xlim = (-4*np.std(X), 4*np.std(X)), ylim = (-2*np.std(Y), 2*np.std(Y)))
        plt.subplot(3,6,plot_num); plot_num += 1
        sns.kdeplot(X, Y, shade=True, cmap="Purples")
        plt.plot([0,0], [-1,1],'k'); plt.plot([-1,1], [0,0],'k')
        plt.xlim([-2*np.std(X), 2*np.std(X)])
        plt.ylim([-2*np.std(Y), 2*np.std(Y)])
        plt.title('-'.join(motif[0]) + ' X ' + '-'.join(motif[1]))
```
