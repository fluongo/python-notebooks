##### Code for working out the nE x T power matrices for a given patient, then performing PC on that resulting dataset

## Plotting the result

```python
import numpy as np
from sklearn.decomposition import PCA, FastICA, SparsePCA
import sys
import pandas as pd
sys.path.append('C:\\python_modules\\subnets-ecog')
import ecogRecordingParams as recParams
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import DataFrame
from scipy.ndimage.filters import gaussian_filter1d as gauss_1d
import scipy.stats as stats
import collections
import datetime
import os


patientID = 'EC63'
data = np.load('C:\\data_' + patientID + '.npy')
```

```python
def perform_PCA_ICA(X):
    # PCA
    nPCs = 10
    pca = PCA(n_components=nPCs)           #X = X.T # Form needs to be nDimension x T
    pc = pca.fit_transform(X)  # Reconstruct signals
    pcEigV = pca.explained_variance_ratio_    
    #plt.plot(pcEigV)

    #ICA
    ica = FastICA(n_components=5)
    ic = ica.fit_transform(X)
    
    return pc, ic

def group_regions(pc, patientID):
    output = recParams.electrode_locs_fq_bands(patientID)
    pcDataFrame = np.zeros((len(output['brainRegions'].keys()), pc.shape[1]))
    kk = 0
    for region in output['brainRegions'].keys():
        pcDataFrame[kk,:] = np.mean(pc[output['brainRegions'][region], :], axis = 0)
        kk+=1
    
    pcDataFrame = DataFrame(pcDataFrame, index = output['brainRegions'].keys())
    
    return pcDataFrame

def UTS_2_date(utsTimeAxis):
    dateTimeAxis = np.array([datetime.datetime.fromtimestamp(timepoint) for timepoint in utsTimeAxis])
    return dateTimeAxis
    
def project_template(pcDataFrame):
    refNames_template = ['ant OFC', 'pos OFC', 'ant ST', 'mid ST', 'pos ST', 
                     'inf CIN', 'sup CIN', 'ant INS', 'pos INS', 'Amygdala', 'Hippocampus']
    template = np.zeros((len(refNames_template), pcDataFrame.shape[1]))
    template = DataFrame(template, index = refNames_template)
    for X in pcDataFrame.index:
        
        ########################
        ##### Dealing with exceptions......
        if X == 'OFC':
            X_old = 'OFC'
            for X in ['pos OFC', 'ant OFC']:
                template.loc[X,:] = pcDataFrame.loc[X_old,:]
            continue
        if X == 'INS':
            X_old = 'INS'
            for X in ['pos INS', 'ant INS']:
                template.loc[X,:] = pcDataFrame.loc[X_old,:]
            continue
        ######################################    
        
        if X in refNames_template:
            template.loc[X,:] = pcDataFrame.loc[X,:]
    return template

# Get rid of NaNs/outliers
X = data[()]['beta']
X[X<-np.percentile(X,0.1)] = -np.percentile(X,0.1) # Clip lower
X[X>np.percentile(X,99.99)] = np.percentile(X,99.99) # Clip upper

pc, ic = perform_PCA_ICA(X)

plt.figure(figsize = (22, 10))
plt.subplot(1,3,1)
sns.heatmap(DataFrame(pc), linewidths =0.0001, cbar = False); plt.title('ALL ELECTRODES')

plt.subplot(1,3,2)
test = group_regions(pc, patientID); sns.heatmap(test, cbar = False); plt.title('Downsample by REGION')

plt.subplot(1,3,3)
template = project_template(test); sns.heatmap(template, cbar = False); plt.title('Projected onto COMMON TEMPLATE')

#sns.clustermap(test, col_cluster=False)
#sns.clustermap(template, col_cluster = False)
```
