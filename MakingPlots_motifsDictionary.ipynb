## Checking The plots

```python
%matplotlib inline
import seaborn as sns

def plot_coords(coords, vals):
    X = np.zeros([len(refNames_template), len(refNames_template)])
    
    for kk, V in zip(coords, vals):
        X[tuple(kk)] = V
        X[tuple(sort(kk))] = V   
    template = DataFrame(X, index = refNames_template, columns = refNames_template)
    sns.heatmap(template, annot = False, fmt = '.2n', square = True, xticklabels = False,  cbar = False, center = 0); 
    
band = 'theta'
nM = '3'
data_fn = compSpec + os.sep + 'output_data' + os.sep + 'motifs_' + band + '.npy'
data = np.load(data_fn)

unique_M = data[()][nM]['motifs']
unique_V = data[()][nM]['values'] 
counts = data[()][nM]['counts']

plt.figure(figsize = [20,4])
kk = 1
for X in data[()].keys()[0:4]:
    plt.subplot(1,4,kk); kk+=1
    countsT = data[()][X]['counts']
    plt.hist(countsT, bins = np.arange(max(countsT)+2)+0.5); plt.yscale('log'); plt.title(X + ' region Motifs')
    
# Code below will plot the top 10 non-diagonal

indices = sorted(range(len(counts)), key=lambda i: counts[i])[-900:]
plt.figure(figsize = [18,18])
kk = 1
for X in reversed(indices):
    if sum(diff(array(unique_M[X]))) != 0: # This enrichs for at least one non-diagonal interaction
        subplot(3,3,kk); kk+=1
        plot_coords(unique_M[X], unique_V[X])
        plt.title('occurred ' + str(counts[X]) + ' times')
    if kk == 10:
        break
```

```python
#Just the plain old top 10

indices = sorted(range(len(counts)), key=lambda i: counts[i])[-9:]

plt.figure(figsize = [18,18])
kk = 1

for X in reversed(indices):
    subplot(3,3,kk); kk+=1
    plot_coords(unique_M[X], unique_V[X])
    plt.title('occurred ' + str(counts[X]) + ' times')
```

## Making a plot of the frequency for the top 100 motifs

```python
#band = 'theta'
#data_fn = compSpec + os.sep + 'output_data' + os.sep + 'disV_motifs_' + band + '.npy'
#data = np.load(data_fn)

def make_composite_firstN_motif(unique_M, unique_V, counts, firstN, noD = False):
    if noD == False:
        indices = sorted(range(len(counts)), key=lambda i: counts[i])[-firstN:]
        X = np.zeros([len(refNames_template), len(refNames_template)])
        for i in indices:
            temp = np.zeros([len(refNames_template), len(refNames_template)])
            for kk, V in zip(unique_M[i], unique_V[i]):
                temp[tuple(kk)] = abs(V)
                temp[tuple(sort(kk))] = abs(V)   
            X += temp
    else:
        indices = sorted(range(len(counts)), key=lambda i: counts[i])[-firstN*100:]
        X = np.zeros([len(refNames_template), len(refNames_template)])
        counter = 1
        for i in reversed(indices):
            if sum(diff(array(unique_M[i]))) != 0: # This enrichs for at least one non-diagonal interaction
                temp = np.zeros([len(refNames_template), len(refNames_template)])
                for kk, V in zip(unique_M[i], unique_V[i]):
                    temp[tuple(kk)] = abs(V)
                    temp[tuple(sort(kk))] = abs(V)
                X += temp; 
                counter+=1
            if counter == firstN:
                break
        
        
    return X

for X in data[()].keys():
    countsT = data[()][X]['counts']
    plt.scatter(int(X)*np.ones_like(countsT) + 0.5*np.random.rand(size(countsT)) - 0.25, countsT)
    plt.xlabel('motif size'); plt.ylabel('number of occurences')
    


for nM in data[()].keys():
    plt.figure()
    temp = make_composite_firstN_motif(data[()][nM]['motifs'], data[()][nM]['values'], data[()][nM]['counts'], 200)
    total_counts = DataFrame(temp, index = refNames_template, columns = refNames_template)
    sns.heatmap(total_counts, annot = True, fmt = '.2n', square = True,  cbar = True); 
    plt.title(str(nM))
```

### Doing the same as above except across frequency bands for a set number of motif size....

```python
nM = '5'

plt.figure(figsize = [18, 12])

kk = 1
for band in ['delta', 'theta', 'alpha', 'beta', 'gamma', 'gammaHi']:
    data_fn = compSpec + os.sep + 'output_data' + os.sep + 'disV_motifs_' + band + '.npy'
    data = np.load(data_fn)
    plt.subplot(2,3,kk); kk+=1
    temp = make_composite_firstN_motif(data[()][nM]['motifs'], data[()][nM]['values'], data[()][nM]['counts'], 200)
    total_counts = DataFrame(temp, index = refNames_template, columns = refNames_template)
    sns.heatmap(total_counts, annot = True, fmt = '.2n', square = True,  cbar = False); 
    plt.title(band.upper() + 'occurences/ top 200 motifs')
```
